```python
score = pd.read_csv("student_scores.csv")
score.shape
```

```python
score.head()
```

```python
score.describe()
```

```python
score.plot(x='Math Score',y='Science Score',style='o')
plt.show()
```

```python
X=score.iloc[:,:-1].values
y=score.iloc[:,1].values
```

```python
X = score[['Math Score']].values
y = score['Science Score'].values
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X,y)
```

```python
print(reg.intercept_)
```

```python
print(reg.coef_)
```

```python
y_pred=reg.predict(X)
```

```python
df=pd.DataFrame({'Actual':y,'Predicted':y_pred})
df
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error(y,y_pred)
```

```python
print('MSE: %.2f' % mean_squared_error(y,y_pred))
```

```python
print('R2_score: %.2f' % r2_score(y,y_pred))
```

```python
plt.scatter(X,y, color='black')
plt.plot(X,y_pred,color='blue', linewidth=2)
plt.show()
```
