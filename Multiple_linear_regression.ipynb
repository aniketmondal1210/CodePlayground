```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from google.colab import files
```

Dataset: It is having investment and profit of 50 startup at USA

```python
# Upload the dataset
upload = files.upload()
```

```python
# Load the dataset
Score = pd.read_csv('startup_data.csv')
```

```python
# Prepare the data
X = Score.drop(['Profit', 'State'], axis=1)
y = Score['Profit']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Create and train the linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = regressor.predict(X_test)
```

```python
# Print the results
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df)
```

```python
# Evaluate the model
print('Coefficients: \n', regressor.coef_)
print('Mean squared error: %.2f' % mean_squared_error(y_test, y_pred))
print('R^2 Score: %.2f' % r2_score(y_test, y_pred))
```

```python
# Plot the results (modified)
plt.scatter(X_test.iloc[:, 0], y_test, color='black', label='Actual')  # Use the first column
plt.scatter(X_test.iloc[:, 0], y_pred, color='blue', label='Predicted')  # Use the first column
plt.xlabel(X_test.columns[0])  # Set the x-axis label
plt.ylabel('Profit')  # Set the y-axis label
plt.legend()  # Add a legend
plt.show()
```
